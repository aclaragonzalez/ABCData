# **Importando os dados**

```python
# importando a biblioteca Pandas
import pandas as pd
```

```python
# leitura do arquivo json que contem os dados
dados = pd.read_csv('./Dados/alura-voz-vf.csv')
```

# **Análises gráficas**

```python
# importando as bibliotecas
import seaborn as sns
```

```python
# estabelecendo o tema dos gráficos a serem plotados
sns.set_theme(style="darkgrid", palette="Paired", font_scale=1.2, rc={"figure.figsize":(13, 6)})
```

```python
# plotando gráfico de Cancelado
sns.displot(dados, x='Cancelado', height=5, aspect=2).set(title='Distribuição da variável Cancelado')
```

```python
# plotando o gráfico de Gênero
sns.displot(dados, x='Gênero', hue='Cancelado', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Gênero por Cancelado')
```

Ao analisar o gráfico acima, não observamos qualquer distinção significativa entre o gênero Masculino e Feminino em relação ao cancelamento do serviço. Isso porque a proporção de pessoas que cancelaram entre os gêneros é a mesma.

```python
# plotando gráfico de Serviço de Internet
sns.displot(dados, x='Serviço.Internet', hue='Cancelado', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Serviço de Internet por Cancelado')
```

Neste gráfico pode ser observado que os clientes que possuem Fibra Óptica como Serviço de Internet apresentam uma maior proporção de cancelamento. No mesmo sentido, os clientes com internet do tipo DSL apresentam maior proporção de cancelamento do que os que não pagam pelo Serviço de Internet.

```python
# plotando gráfico de Seguraça Online
sns.displot(dados, x='Segurança.Online', hue='Cancelado', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Segurança Online por Cancelado')
```

```python
# plotando o gráfico de Backup Virtual
sns.displot(dados, x='Backup.Virtual', hue='Cancelado', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Backup Virtural por Cancelado')
```

```python
# plotando o gráfico de Proteção de Dispositivo
sns.displot(dados, x='Proteção.Dispositivo', hue='Cancelado', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Proteção de Dispositivo por Cancelado')
```

```python
# plotando o gráfico de Suporte Técnico
sns.displot(dados, x='Suporte.Técnico', hue='Cancelado', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Suporte Técnico por Cancelado')
```

Nos gráficos acima se observa uma maior proporção de cancelamento por quem não possui os serviços adicionais (Segurança Online, Backup Virtual, Proteção de Dispositivo e Suporte Técnico).

```python
# plotando gráfico de Contrato
sns.displot(dados, x='Contrato', hue='Cancelado', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Contrato por Cancelado')
```

Neste gráfico se observa que os clientes que fazem planos mais longíquos apresentam uma menor taxa de cancelamento do que os que possuem contrato mensal.

```python
# plotando gráfico de Método de Pagamento
sns.displot(dados, x='Método.Pagamento', hue='Cancelado', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Método de Pagamento por Cancelado')
```

O gráfico acima mostra que os clientes que utilizam o Método de Pagamento de cheque eletrônico têm uma maior propensão de cancelar os serviços.

```python
# plotando gráfico de Idoso
sns.displot(dados, x='Idoso', hue='Cancelado', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Idoso por Cancelado')
```

Embora a quantidade de idosos compreenda uma parcela relativamente pequena do total de clientes, a partir do gráfico acima pode ser observado que uma grande parte (quase 50%) dos idosos tendem a cancelar os serviços.

```python
# plotando gráfico de Idoso
sns.displot(dados, x='Idoso', hue='Suporte.Técnico', multiple='stack', height=5, 
            aspect=2).set(title='Distribuição da variável Idoso por Suporte Técnico')
```

A maior parte dos idosos não possuem o serviço de Suporte Técnico, que pode ser um indicativo da alta taxa de cancelamento por esse grupo de clientes, visto que os idosos possuem uma maior dificuldade com as tecnologias.

```python
sns.boxplot(x='Vigência', y='Cancelado', data=dados).set(title='Boxplot de Vigência por Churn')
```

A partir da análise do boxplot acima se observa que os clientes com maiores tempos de contrato tendem a continuar utilizando o serviço.

```python
sns.boxplot(x='Valor.Mensal', y='Cancelado', data=dados).set(title='Boxplot de Valor Mensal por Churn')
```

Pelo boxplot acima pode ser observado que os clientes com maior propensão ao cancelamento dos serviços são aqueles cujos valores mensais são maiores.

```python
corr = dados.corr()
```

```python
sns.heatmap(corr, cmap='YlGnBu')
```

A partir do HeatMap se observa uma forte correlação entre o tempo de Vigência do contrato e as variáveis Valor Total e Valor Diário.
